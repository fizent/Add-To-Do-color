```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from google.colab import files

# بارگیری داده‌های CIFAR10
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0


uploaded = files.upload()



# تعریف تکنیک افزایش داده با تنظیمات پیشرفته
datagen = ImageDataGenerator(
    rotation_range=40,          # چرخش تصادفی
    width_shift_range=0.2,      # تغییر عرض
    height_shift_range=0.2,     # تغییر ارتفاع
    shear_range=0.2,            # کشش تصادفی
    zoom_range=0.2,             # زوم تصادفی
    horizontal_flip=True,       # چرخش افقی
    fill_mode='nearest'         # نحوه پر کردن مناطق خالی
)
datagen.fit(x_train)

# ساخت مدل بهبود یافته با استفاده از BatchNormalization، Dropout و لایه‌های عمیق‌تر
model = tf.keras.models.Sequential([
    # بلوک اول
    tf.keras.layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),

    # بلوک دوم
    tf.keras.layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),

    # بلوک سوم
    tf.keras.layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),

    # بخش Fully Connected
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10, activation='softmax')
])

# کامپایل مدل
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# تعریف callbacks برای بهبود روند آموزش
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1),
    ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True, verbose=1)
]

# آموزش مدل با استفاده از داده‌های افزایش یافته
history = model.fit(datagen.flow(x_train, y_train, batch_size=64),
                    epochs=50,
                    validation_data=(x_test, y_test),
                    callbacks=callbacks)

# تابع برای پیش‌بینی چند تصویر
def predict_multiple_images(image_paths):
    class_names = ["Airplane", "Car", "Bird", "Cat", "Deer", "Dog", "Frog", "Horse", "Boat", "Truck"]
    for img_path in image_paths:
        # بارگذاری و پیش‌پردازش تصویر
        img = image.load_img(img_path, target_size=(32, 32))
        img_array = image.img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        # پیش‌بینی با مدل
        predictions = model.predict(img_array)
        predicted_class = class_names[np.argmax(predictions[0])]

        # نمایش تصویر و پیش‌بینی
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"پیش‌بینی: {predicted_class}")
        plt.show()

# مثال استفاده: لیست مسیر تصاویر (مسیرهای تصاویر خود را جایگزین کنید)
image_paths = ['/content/airplan.jfif', '/content/car.png', '/content/dog.jfif', "/content/truck.jfif", "/content/سگ 2.png", "/content/گربه 2.jfif", "/content/اهو.png", "/content/قورباغه.png"]
predict_multiple_images(image_paths)
```

```python
import torch
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
from diffusers import StableDiffusionPipeline

# بارگذاری مدل‌های CLIP و Stable Diffusion
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

stable_diffusion = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
stable_diffusion = stable_diffusion.to("cpu")

# تابع برای دریافت امبدینگ تصویر و متن
def get_embedding(image_path, text_prompt):
    # پردازش تصویر
    image = Image.open(image_path).convert("RGB")
    input_image = clip_processor(images=image, return_tensors="pt")
    image_embedding = clip_model.get_image_features(**input_image)

    # پردازش متن
    input_text = clip_processor(text=[text_prompt], return_tensors="pt", padding=True)
    text_embedding = clip_model.get_text_features(**input_text)

    # نرمال‌سازی امبدینگ‌ها
    image_embedding = image_embedding / image_embedding.norm(dim=-1, keepdim=True)
    text_embedding = text_embedding / text_embedding.norm(dim=-1, keepdim=True)

    return image_embedding, text_embedding

# ترکیب امبدینگ‌های تصویر و متن
def combine_embedding(image_embedding, text_embedding, alpha=0.5):
    combined_embedding = alpha * image_embedding + (1 - alpha) * text_embedding
    combined_embedding = combined_embedding / combined_embedding.norm(dim=-1, keepdim=True)
    return combined_embedding

# تولید تصویر با Stable Diffusion
def generate_image(prompt):
    image = stable_diffusion(prompt).images[0]
    return image

# تست نمونه‌ها
examples = [
    {"image_path": "/content/car.png", "text_prompt": "car in city"}
]

for example in examples:
    image_embedding, text_embedding = get_embedding(example["image_path"], example["text_prompt"])
    combined_embedding = combine_embedding(image_embedding, text_embedding, alpha=0.6)

    # تولید تصویر و ذخیره آن
    generated_image = generate_image(example["text_prompt"])
    image_path = f"generated_{example['text_prompt'].replace(' ', '_')}.png"
    generated_image.save(image_path)

    print(f"تصویر ذخیره شد: {image_path} ✅")
```
